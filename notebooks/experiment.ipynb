```python
import pandas as pd
import numpy as np
import ast

cora = pd.read_csv('../results_csv/Cora_accuracy_results.csv')
enzymes = pd.read_csv('../results_csv/ENZYMES_accuracy_results.csv')
imdb = pd.read_csv('../results_csv/IMDB-BINARY_accuracy_results.csv')
```

```python
def calc_stat(df):
    df['Values'] = df['Values'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

    # Calculate rounded median and standard deviation
    df['Median_Value'] = df['Values'].apply(lambda x: round(np.median(x), 3))
    df['Std_Deviation'] = df['Values'].apply(np.std)

    # Format as "Median ± Std_Deviation"
    df['Median ± Std'] = df.apply(lambda row: f"{row['Median_Value']} ± {row['Std_Deviation']:.2f}", axis=1)

    return df

# Apply to your DataFrames
cora = calc_stat(cora)
enzymes = calc_stat(enzymes)
imdb = calc_stat(imdb)
```
